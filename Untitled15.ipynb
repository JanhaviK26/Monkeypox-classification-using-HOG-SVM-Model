```python
import os
import cv2
data = []
sorted_dir=sorted(os.listdir('/content/drive/MyDrive/newtrain11/'))
print(sorted_dir)
img = np.array(cv2.imread('sorted_dir'))
```

```python
data=[]
entry=sorted(glob.glob(os.path.join("/content/drive/MyDrive/newtrain11/",'*.png')))
for entries in entry:
  print(entries)
  image = np.array(mpimg.imread(entries))
  resized_img = cv2.resize(image, (64,32))
  fd = hog(resized_img,channel_axis=-1)
  data.append(fd)
```

```python
train_list = []
for i in range(len(data)) :
    if i < 108 :
        train_list.append("Monkeypox")
    else :
        train_list.append("nonMonkeypox")
print(train_list)
print(len(train_list))
```

```python
test_Monkeypox = []
for entries in glob.glob("/content/drive/MyDrive/Disease Detection (Internship)/monkeypox/archive/Original Images/Original Images/MP/*.png"):
    img = np.array(mpimg.imread(entries))
    resized_img = cv2.resize(img, (64, 32))
    fd = hog(resized_img,channel_axis=-1)
    test_dict = {'data': fd, 'lable': 'Monkeypox'}
    test_Monkeypox.append(test_dict)

test_nonMonkeypox = []
for entries in glob.glob("/content/drive/MyDrive/Disease Detection (Internship)/monkeypox/archive/Original Images/Original Images/NMP/*.png"):
    img = np.array(mpimg.imread(entries))
    resized_img = cv2.resize(img, (64,32))
    fd = hog(resized_img,channel_axis=-1)
    test_dict = {'data': fd, 'lable': 'nonMonkeypox'}
    test_nonMonkeypox.append(test_dict)

test_data = test_Monkeypox + test_nonMonkeypox
print(len(test_data))
```

```python
test_features = []
test_lables = []
for i in test_data:
    test_lables.append(i['lable'])
    test_features.append(i['data'])
print(len(test_features))
print(test_lables)
```

```python
from sklearn import svm

clf = svm.SVC(kernel='linear')
clf.fit(data,train_list)
pred = clf.predict(test_features)
accuracy = clf.score(test_features,test_lables)
print('Accuracy: ',accuracy)
print(metrics.confusion_matrix(test_lables, pred))
print('F1 Score: ',metrics.f1_score(test_lables, pred,average='macro'))
print(classification_report(test_lables, pred))
```
